```
disney = pd.read_csv('https://cdn.touringplans.com/datasets/metadata.csv', 
                     parse_dates=['DATE'])
```

```
# putting values from string into float form without '%' for use in numpy
disney['inSession'] = [int(x.strip('%')) for x in disney['inSession']]
disney['inSession_wdw'] = [int(x.strip('%')) for x in disney['inSession_wdw']]
disney['inSession_sqrt_WDW'] = [int(x.strip('%')) for x in disney['inSession_sqrt_WDW']]
disney['inSession_Central_FL'] = [int(x.strip('%')) for x in disney['inSession_Central_FL']]
disney['inSession_Drive1_FL'] = [int(x.strip('%')) for x in disney['inSession_Drive1_FL']]
disney['inSession_Drive2_FL'] = [int(x.strip('%')) for x in disney['inSession_Drive2_FL']]
disney['inSession_Florida'] = [int(x.strip('%')) for x in disney['inSession_Florida']]
disney['INSESSION_PLANES'] = [int(x.strip('%')) for x in disney['INSESSION_PLANES']]
```

```
# split into years
disney_2012 = disney[disney['YEAR']==2012]
disney_2013 = disney[disney['YEAR']==2013]
disney_2014 = disney[disney['YEAR']==2014]
disney_2015 = disney[disney['YEAR']==2015]
disney_2016 = disney[disney['YEAR']==2016]
disney_2017 = disney[disney['YEAR']==2017]
disney_2018 = disney[disney['YEAR']==2018]
disney_2019 = disney[disney['YEAR']==2019]
```
